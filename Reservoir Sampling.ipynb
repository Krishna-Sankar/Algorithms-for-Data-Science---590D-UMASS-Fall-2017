QUESTION:

Implement reservoir sampling when reservoir has size 1. Let the items from 1 to 100 appear one by one.
1. reservoiriter1.py -Report the item sampled in one run of the algorithm.
2. reservoiriter1000.py -Repeat the algorithm for 1000 times and plot the number of times each element is selected.
3. reservoiriter10000.py -Repeat the algorithm for 10000 times and plot the number of times each element is selected.
4. reservoiriter100000.py -Repeat the algorithm for 100000 times and plot the number of times each element is selected.

item_array[] - stream of elements from 1,2,3,......100
n - reservoir size
reservoir - reservoir element
selected - selected element with probability 1/j for jth element

```python
#reservoiriter1.py
import random
item_array=[0]*100
n=1
for i in range(1,100):
    item_array[i] = i
reservoir=item_array[n]
for j in range(n+1, 101):
        selected = random.randrange(1, j, 1)
        if selected != reservoir:
            reservoir = selected
print " Item Selected after 1 Iteration is ",selected
```

count[] - array storing count of how many times each item, in the stream has been selected
iter - iterator for while loop used for running the algorithm multiple times

Bar Graph has been plotted as output 
Count array printed to give an idea of what values has been plotted.
